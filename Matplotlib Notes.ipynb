```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 5, 11)   #linearly spaced 05 11
y = x ** 2

plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()   #

plt.subplot(1,2,1)    #sub
plt.plot(x, y, 'b') # More on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])  #left&bottom axis,figwidthheight
axes.plot(x,y)                          #plot
axes.set_xlabel('hahaha')
axes.set_ylabel('hohoho')    #fig
axes.set_title('hhh')
```

```python
#
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('big')
axes2.plot(y,x)
axes2.set_title('small')    #

```

```python
#
fig,axes = plt.subplots(nrows=1,ncols=2)  #nrows*ncols 1*2

axes[0].plot(x,y)  #plot
axes[0].set_title('hahaha')   #
axes[1].plot(y,x)
axes[1].set_xlabel('hihi')

plt.tight_layout() #deal with overlapping plots
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2)) #m*n  in inches 

axes[0].plot(x,y)
```

```python
#save a figure
fig.savefig('hahahahhaha.png',dpi=100) #dpi 
#
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**3,color='#FF8C00',label='X squared',
linewidth=0.8,alpha=0.4,
linestyle='--',marker='o',markersize='4',
markerfacecolor='black')                        
#RGB hexcodes/name(purple) 
#
#,alpha,
#markerx'o','1','+'),

ax.plot(y,x**2,color='blue',lw=2)
#

ax.legend(loc=0) #,loc=0-10,0=best,loc=(a,b)

ax.set_xlim([0,10])
ax.set_ylim([0,20]) #xy
```
