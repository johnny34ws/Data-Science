```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 5, 11)   #linearly spaced 0到5 11个数字
y = x ** 2

plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()   #最好写一下，用来显示

plt.subplot(1,2,1)    #sub图
plt.plot(x, y, 'b') # More on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])  #left&bottom axis的位置,fig本身的width和height
axes.plot(x,y)                          #更面向对象，并plot
axes.set_xlabel('hahaha')
axes.set_ylabel('hohoho')    #fig下不同的设置方法
axes.set_title('hhh')
```

```python
#嵌套图
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('big')
axes2.plot(y,x)
axes2.set_title('small')    #一图两函数

```

```python
#多图
fig,axes = plt.subplots(nrows=1,ncols=2)  #nrows*ncols的坐标图 下为1*2

axes[0].plot(x,y)  #给某个plot画图
axes[0].set_title('hahaha')   #语法不变
axes[1].plot(y,x)
axes[1].set_xlabel('hihi')

plt.tight_layout() #deal with overlapping plots
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2)) #m*n的图片 以及 大小in inches 

axes[0].plot(x,y)
```

```python
#save a figure
fig.savefig('hahahahhaha.png',dpi=100) #dpi为像素 
#存在本程序的目录下
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**3,color='#FF8C00',label='X squared',
linewidth=0.8,alpha=0.4,
linestyle='--',marker='o',markersize='4',
markerfacecolor='black')                        
#RGB hexcodes/name(purple) 
#给函数起名字
#函数粗细,透明度alpha,线样式（可查）
#marker显示x的每个点（'o','1','+'),以及其大小和颜色

ax.plot(y,x**2,color='blue',lw=2)
#一图多线

ax.legend(loc=0) #显示函数名字,loc=0-10,0=best,其他个性化；loc=(a,b)对应坐标

ax.set_xlim([0,10])
ax.set_ylim([0,20]) #限制x和y的大小
```
